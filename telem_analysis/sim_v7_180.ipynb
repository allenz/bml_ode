```python
import matplotlib.pyplot as plot
%matplotlib inline
import numpy as np
import pandas as pd
import base
```

```python
# Extended telemetry with 180 degree offset
t = base.process('sim_v7_180_long.csv')
base.setPlot(plot)
```

```python
base.plotPAT(t, [6000,9000])
```

# Position and Angle

```python
# The VelociROACH moves 0.1 m/s at 1.5 Hz; turns 180 degrees in 10 s
plot.plot(t.time, t.x, 'r', t.time, t.y, 'g', t.time, t.g, 'b', t.time, t.AngleZ, 'c')
plot.xlabel('Time (ms)')
plot.ylabel('Position (m) / Angle (rad)')
plot.legend(['x','y','Yaw', 'Est yaw'], loc=2)
```

```python
t['VelX'] = t.x.diff()
t['VelY'] = t.y.diff()
plot.plot(t.time, np.sqrt(np.square(t[['VelX','VelY']]).sum(axis=1)), 'b')
```

```python
# Turns left
plot.plot(t.x, t.y, 'b')
plot.xlabel('x (m)')
plot.ylabel('y (m)')
plot.axis('equal')
```

```python
# Complicated dynamics: low-frequency changes, not periodic
# Est Roll from gyro drifts.. should remain near 0 because robot is on the ground
plot.plot(t.time, t.a, 'r', t.time, t.b, 'b', t.time, t.g/10, 'k')#, t.time, t.AngleY, 'g')
plot.xlabel('Time (ms)')
# plot.xlim([3000,6000])
plot.ylabel('Angle (rad)')
plot.legend(['Pitch','Roll', 'Yaw/10'], loc=1)#,'Est Roll'
```

```python
# Complicated dynamics: low-frequency changes, not periodic
# Est Roll from gyro drifts.. should remain near 0 because robot is on the ground

plot.plot(t.time, t.GyroY, 'r', t.time, t.GyroZ, 'b', t.time, t.g/10, 'k')#, t.time, t.AngleY, 'g')
plot.xlabel('Time (ms)')
plot.ylabel('Angle (rad)')
plot.legend(['Gyro Pitch','Gyro Roll', 'Yaw/10'], loc=1)#,'Est Roll'
```

```python
# Complicated dynamics: low-frequency changes, not periodic
# Est Roll from gyro drifts.. should remain near 0 because robot is on the ground

plot.plot(t.time, t.GyroY, 'r', t.time, t.GyroZ, 'b', t.time, t.g/10, 'k')#, t.time, t.AngleY, 'g')
plot.xlabel('Time (ms)')
plot.xlim([3000,6000])
plot.ylabel('Angle (rad)')
plot.legend(['Gyro Pitch','Gyro Roll', 'Yaw/10'], loc=1)#,'Est Roll'
```

```python
# Complicated dynamics: low-frequency changes, not periodic
# Est Roll from gyro drifts.. should remain near 0 because robot is on the ground

plot.plot(t.time, t.AngleY, 'r', t.time, t.AngleZ, 'b', t.time, t.g/10, 'k')#, t.time, t.AngleY, 'g')
plot.xlabel('Time (ms)')
plot.ylabel('Angle (rad)')
plot.legend(['Pitch (cumsum)','Roll (cumsum)', 'Yaw/10'], loc=2)#,'Est Roll'
```

```python
import pandas as pd
dt = 10/1000 # 10 ms
window = int(np.ceil(1/1.5/dt)) # number of observations per cycle
plot.plot(t.time, pd.rolling_sum(t.GyroX, window), 'r', t.time, pd.rolling_sum(t.GyroY, window), 'b', t.time, t.g/10, 'k')#, t.time, t.AngleY, 'g')
plot.xlabel('Time (ms)')
plot.ylabel('Angle (rad)')
plot.legend(['Pitch (rollsum)','Roll (rollsum)', 'Yaw/10'], loc=1)#,'Est Roll'
```

```python
# Pitches forward when switching off, especially pronounced on right side
plot.plot(t.time, t.a, 'r', t.time, t.b, 'b')
plot.xlim([10e3,13e3])
plot.axhline(color='m')
plot.xlabel('Time (ms)')
plot.xlim([25000,28000])
plot.ylabel('Angle (rad)')
plot.legend(['Pitch','Roll'], loc=2)
```
